##### bloxpot - gráfico de caixa

Permite visualizar os dados com informações de discrepância e distribuição de outliers.

Apresenta estatística de mínimo, máximo, primeiro quartil, segundo quartil, mediana e o terceiro quartil.

#### 1. Boxplot Gráfico - código

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
ax = sns.boxplot(x=tips['total_bill'], palette="light:m_r")
```

Observações sobre o gráfico:

Mínimo = 0 (Desconsiderando Outliers)
Máximo = 40 (Desconsiderando Outliers)
Mediana = 18 
Outliers >40 
Primeiro Quartil = 13
Segundo Quartil = 18
Terceiro Quartil = 24

Dados próximos ao primeiro quartil, portanto assimetria positiva.
